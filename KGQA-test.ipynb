# 加载数据

```python
from pathlib import Path
from llama_index import download_loader

PagedCSVReader = download_loader("PagedCSVReader")

loader = PagedCSVReader(encoding="utf-8")
documents = loader.load_data(file=Path(r'database\relationnew.csv'))
```

## 自带csv文件loader

```python
from langchain.document_loaders.csv_loader import CSVLoader
loader = CSVLoader(file_path=r'database\relationnew.csv')
 
data = loader.load()
```

```python
print(data)
```

# 文档分割

```python
from langchain.text_splitter import CharacterTextSplitter
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
texts = text_splitter.split_documents(data)
```

# 选择嵌入和向量数据库

```python
from langchain.embeddings import OpenAIEmbeddings
embeddings = OpenAIEmbeddings()

from langchain.vectorstores import Chroma
db = Chroma.from_documents(data, embeddings)
```

# 创建索引

```python
retriever = db.as_retriever()
```

# 创建索引问答链

```python
from langchain.llms import OpenAI
from langchain.chains import RetrievalQA
qa = RetrievalQA.from_chain_type(llm=OpenAI(), chain_type="stuff", retriever=retriever,verbose=True)

query = "请根据数据库告诉我爱在黎明破晓前的上映年份"
qa.run(query)
```
